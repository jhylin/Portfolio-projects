```python
import pandas_read_xml as pdx
from pandas_read_xml import fully_flatten
```

```python
#Orphadata: Free access data from Orphanet. 
#© INSERM 1999. Available on http://www.orphadata.org. Data version (XML data version).
#Dataset (.xml file) from http://www.orphadata.org/cgi-bin/epidemio.html
#Latest date of update for the dataset: 7/6/2022
#Creative Commons Attribution 4.0 International

#Read .xml file
df = pdx.read_xml('http://www.orphadata.org/data/xml/en_product9_ages.xml').pipe(fully_flatten)
df.head()
```

```python
df.info()
```

```python
#Dropping columns that are not required for data analysis at this stage
df.drop(columns = ['JDBOR|@date', 
                   'JDBOR|@copyright', 
                   'JDBOR|Availability|Licence|FullName|@lang'], 
        inplace = True)

#Continue to drop further columns not required
df.drop(['JDBOR|@dbserver', 
         'JDBOR|@version', 
         'JDBOR|Availability|Licence|FullName', 
         'JDBOR|Availability|Licence|LegalCode', 
         'JDBOR|Availability|Licence|ShortIdentifier'], 
        axis = 1, inplace = True)

#Drop columns which were just stating English was the language used
df.drop(['JDBOR|DisorderList|Disorder|AverageAgeOfDeathList|AverageAgeOfDeath|Name|@lang', 
         'JDBOR|DisorderList|Disorder|AverageAgeOfOnsetList|AverageAgeOfOnset|Name|@lang', 
         'JDBOR|DisorderList|Disorder|DisorderGroup|Name|@lang', 
         'JDBOR|DisorderList|Disorder|DisorderType|Name|@lang', 
         'JDBOR|DisorderList|Disorder|ExpertLink|@lang', 
         'JDBOR|DisorderList|Disorder|Name|@lang', 
         'JDBOR|DisorderList|Disorder|TypeOfInheritanceList|TypeOfInheritance|Name|@lang'], 
        axis = 1, inplace = True)

df.head()
```

```python
#Check how many columns are left now
df.info()
```

```python
#Check what this column contains
df['JDBOR|DisorderList|Disorder|ExpertLink'].unique()
```

```python
#Drop this column which is referring to links of information from Orphanet for rare diseases only
df.drop(columns = ['JDBOR|DisorderList|Disorder|ExpertLink'], inplace = True)
df.head()
```

```python
#Check how many columns left after removing several before
df.info()
```

```python
#Rename all columns for ease of reading
df.rename(columns = {'JDBOR|DisorderList|@count': 'Disorder list counts', 
                     'JDBOR|DisorderList|Disorder|@id': 'Disorder ID', 
                     'JDBOR|DisorderList|Disorder|AverageAgeOfDeathList|@count': 'Average age of death - counts', 
                     'JDBOR|DisorderList|Disorder|AverageAgeOfDeathList|AverageAgeOfDeath|@id': 'Average age of death ID', 
                     'JDBOR|DisorderList|Disorder|AverageAgeOfDeathList|AverageAgeOfDeath|Name': 'Average age of death', 
                     'JDBOR|DisorderList|Disorder|AverageAgeOfOnsetList|@count': 'Average age of onset counts', 
                     'JDBOR|DisorderList|Disorder|AverageAgeOfOnsetList|AverageAgeOfOnset|@id': 'Average age of onset ID', 
                     'JDBOR|DisorderList|Disorder|AverageAgeOfOnsetList|AverageAgeOfOnset|Name': 'Average age of onset', 
                     'JDBOR|DisorderList|Disorder|DisorderGroup|@id': 'Disorder group ID', 
                     'JDBOR|DisorderList|Disorder|DisorderGroup|Name': 'Disorder group name', 
                     'JDBOR|DisorderList|Disorder|DisorderType|@id': 'Disorder type ID', 
                     'JDBOR|DisorderList|Disorder|DisorderType|Name': 'Disorder type', 
                     'JDBOR|DisorderList|Disorder|Name': 'Disorder name', 
                     'JDBOR|DisorderList|Disorder|OrphaCode': 'Orpha code', 
                     'JDBOR|DisorderList|Disorder|TypeOfInheritanceList|@count': 'Type of inheritance list counts', 
                     'JDBOR|DisorderList|Disorder|TypeOfInheritanceList|TypeOfInheritance|@id': 'Type of inheritance ID', 
                     'JDBOR|DisorderList|Disorder|TypeOfInheritanceList|TypeOfInheritance|Name': 'Type of inheritance'}, 
         inplace = True)

df.head(10)
```

```python
#Create another dataframe with columns: disorder type, disorder name, type of inheritance, 
#average age of onset, average age of death
rd = df[['Disorder type', 'Disorder name', 'Type of inheritance', 'Average age of onset', 'Average age of onset counts', 'Average age of death', 'Average age of death - counts']].copy()
rd.head()
```

```python
#To see all different average age of deaths present in data
rd['Average age of death'].unique()
```

```python
#Observe frequencies of all the categories of average age of death
rd['Average age of death'].value_counts()
```

```python
#Observe frequencies of all the categories of average age of onset
rd['Average age of onset'].value_counts()
```

```python
#Create a filter to find out rare disease disorder type of malformation syndrome only
#with average age of onset as antenatal and average age of death as elderly
selection_mal_syn = (rd['Disorder type'] == 'Malformation syndrome') & (rd['Average age of onset'] == 'Antenatal') & (rd['Average age of death'] == 'elderly')
print(selection_mal_syn)
```

```python
#apply the filter to the dataset rd
a = rd[selection_mal_syn]
a
```

```python
#List the unique disorder name after applying this filter
a['Disorder name'].unique()
```

It turns out that Turner syndrome and Prune belly syndrome are the only two disorders of the malformation syndrome type that have an average age of onset at antenatal period with an average age of death in elderly years. This means these two rare disorders have had relatively long life spans out of all the rare diseases present in the dataset.

```python
#To find out the rare diseases that have the same time of onsets but with earlier deaths than elderly years
#One of the examples
selection_mal_syn_early = (rd['Disorder type'] == 'Malformation syndrome') & (rd['Average age of onset'] == 'Antenatal') & (rd['Average age of death'] == 'early childhood')
b = rd[selection_mal_syn_early]
b
```

```python
b['Disorder name'].unique()
```

Oppositely, there are far more rare disorders (as shown above) with early childhood deaths within the malformation syndrome type, while having the same antenatal onsets as Turner syndrome and Prune belly syndrome.
