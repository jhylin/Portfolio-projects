```python
import pandas as pd
import pandas_read_xml as pdx
from pandas_read_xml import fully_flatten
```

```python
#Orphadata: Free access data from Orphanet. 
#© INSERM 1999. Available on http://www.orphadata.org. Data version (XML data version).
#Dataset (.xml file) from http://www.orphadata.org/cgi-bin/epidemio.html
#Latest date of update for the dataset: 14/6/2022
#Creative Commons Attribution 4.0 International
```

```python
#Import data from Orphanet
df = pdx.read_xml('http://www.orphadata.org/data/xml/en_product4.xml').pipe(fully_flatten)
df.head(5)
```

```python
df.info()
```

```python
#Noted HPO Disorder set status list count = 4245, confirmed with value_counts()
df['JDBOR|HPODisorderSetStatusList|@count'].value_counts()
```

```python
#Decided to also remove rest of the ID columns as not needed for analysis for now 
#except human phenotype ontology (HPO) ID
df.drop(['JDBOR|HPODisorderSetStatusList|HPODisorderSetStatus|Disorder|HPODisorderAssociationList|HPODisorderAssociation|@id', 
         'JDBOR|HPODisorderSetStatusList|HPODisorderSetStatus|Disorder|HPODisorderAssociationList|HPODisorderAssociation|DiagnosticCriteria|@id', 
         'JDBOR|HPODisorderSetStatusList|HPODisorderSetStatus|Disorder|HPODisorderAssociationList|HPODisorderAssociation|HPOFrequency|@id'],
       axis = 1, inplace = True)
```
