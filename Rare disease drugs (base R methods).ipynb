```r
#.csv file needs to be saved only as comma-separated values type so it'll be read as a dataframe 
#Not .csv UTF-8 type, which will be read as list

#Read .csv file
df <- read.csv("Rare_diseases.csv", sep = ",", header = TRUE)

#Show first 5 rows of the dataframe
head(df, 5)
```

```r
#Change column name of Generic.Name by using index
colnames(df)[1] <- c("Generic Name")

#Change column names using old column names
colnames(df)[which(names(df) == "Trade.Name")] <- "Trade Name"
colnames(df)[which(names(df) == "Date.Designated")] <- "Date Designated"
colnames(df)[which(names(df) == "Orphan.Designation")] <- "Orphan Designation"
```

```r
head(df,5)
```

```r
#Check .csv file is read into a dataframe and check numbers of columns and rows in total
print(is.data.frame(df))
print(ncol(df))
print(nrow(df))
```

```r
#Remove column FDA.Orphan.Approval.Status 
df <- subset(df, select = -c(FDA.Orphan.Approval.Status))
head(df, 5)
```

```r
#Drop these two columns as this dataset was for approved rare disease drugs only and 
#CF Grid Key was for FDA office use only
df <- subset(df, select = -c(Orphan.Designation.Status, CF.Grid.Key))

#Drop address columns as not needed for now
df <- subset(df, select = -c(Sponsor.Address.1, Sponsor.Address.2, Sponsor.City, Sponsor.Zip))
head(df, 5)
```

```r
#Tidy up more column names
colnames(df)[which(names(df) == "Approved.Labeled.Indication")] <- "Approved labelled indication"
colnames(df)[which(names(df) == "Marketing.Approval.Date")] <- "Marketing approval date"
colnames(df)[which(names(df) == "Exclusivity.End.Date")] <- "Exclusivity end date"
colnames(df)[which(names(df) == "Exclusivity.Protected.Indication....Shown.for.approvals.from.Jan..1..2013..to.the.present.")] <- "Exclusivity protected indication"
colnames(df)[which(names(df) == "Sponsor.Company")] <- "Sponsor company"
colnames(df)[which(names(df) == "Sponsor.State")] <- "Sponsor state"
colnames(df)[which(names(df) == "Sponsor.Country")] <- "Sponsor country"
head(df, 5)
```

```r
#Check numbers of columns and rows after removing some of the columns
print(ncol(df))
print(nrow(df))
```

```r
#Show last five rows of the dataframe
tail(df, 5)
```

```r
#Remove row number 1058 as this was only an explanatory caption for exclusivity protected indication
df <- df[-c(1058), ]
tail(df, 5)
```

```r
#Check countries involved in developing rare disease drugs 
unique(df["Sponsor country"])
```

```r
#Find out the frequencies of countries involved in developing rare disease drugs
table1 = table(df$"Sponsor country")
table1
```

```r
#Convert above table into a new dataframe
new_df <- data.frame(rbind(table1))
new_df
```

```r
#Transpose the column and row to make it easier to read
df_transpose = t(new_df)
df_transpose
```

```r
#Add column name for frequency counts
colnames(df_transpose) <- c('Frequency')
#Add column name for the countries listed
df_transpose <- cbind(Country = rownames(df_transpose), df_transpose)
#Add index to rows 
rownames(df_transpose) <- 1:nrow(df_transpose)

df_transpose
```

```r
#Change above matrix into a dataframe
df_new = as.data.frame.matrix(df_transpose)
df_new
```

```r
#Plot horizontal bar graph for all the countries involved in rare disease drug development by using above dataframe
library(ggplot2)
#Adjust size of the bar graph
options(repr.plot.width = 10, repr.plot.height = 10)
#Plot using ggplot
ggplot(df_new, aes(x = Country, y = Frequency)) + 
geom_bar(stat = "identity") +
#Change to horizontal bar graph
coord_flip()
```

From the above horizontal graph, United States, Switzerland, Canada and Sweden were the top 3 countries involved in rare disease drug development (Canada and Sweden both took the third place together).
